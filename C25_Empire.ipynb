```python
# CELL 2: LOAD ALL BUILD HISTORY
import hashlib

TRUTH_DIR = '/sdcard/TERMUX_TRUTH'
commands = []
sh_files = []
py_files = []

if os.path.exists(TRUTH_DIR + '/TRUTH_MANIFEST.json'):
    with open(TRUTH_DIR + '/TRUTH_MANIFEST.json') as f:
        truth = json.load(f)
    commands = truth.get('top_commands', [])
    sh_files = truth.get('sh_files_list', [])
    py_files = truth.get('py_files_list', [])

# Also read full history
if os.path.exists(TRUTH_DIR + '/FULL_HISTORY.txt'):
    with open(TRUTH_DIR + '/FULL_HISTORY.txt', errors='ignore') as f:
        for line in f:
            line = line.strip()
            if line and not line.startswith('[') and not line.startswith('='):
                commands.append(line)

commands = list(set(commands))
print(f'Unique commands loaded: {len(commands):,}')
print(f'Bash scripts loaded   : {len(sh_files):,}')
print(f'Python scripts loaded : {len(py_files):,}')
print()
print('Sample commands:')
for cmd in commands[:10]:
    print(f'  {cmd}')
```

```python
# CELL 3: RUN ALL CONFIRMED EXECUTED SCRIPTS
import subprocess

SKIP = ['node_modules','site-packages','venv','.git','nvm.sh','flutter','thrift','pnpm','cassandra']
TIMEOUT = 10

passed = []
failed = []
skipped = []

confirmed_file = TRUTH_DIR + '/CONFIRMED_EXECUTED.txt'
if os.path.exists(confirmed_file):
    with open(confirmed_file) as f:
        lines = f.readlines()

    scripts = []
    for line in lines:
        parts = line.strip().split()
        if len(parts) >= 2:
            scripts.append(parts[1])

    print(f'Running {len(scripts)} confirmed-executed scripts...')
    for i, script in enumerate(scripts):
        if not os.path.exists(script):
            skipped.append(script)
            continue
        if any(s in script for s in SKIP):
            skipped.append(script)
            continue
        try:
            runner = 'python3' if script.endswith('.py') else 'bash'
            r = subprocess.run([runner, script], capture_output=True, timeout=TIMEOUT)
            if r.returncode == 0:
                passed.append(script)
                print(f'  ✓ [{i+1}] {os.path.basename(script)}')
            else:
                failed.append(script)
                print(f'  ✗ [{i+1}] {os.path.basename(script)}')
        except subprocess.TimeoutExpired:
            skipped.append(script)
            print(f'  ⏱ [{i+1}] {os.path.basename(script)} TIMEOUT')
        except Exception as e:
            failed.append(script)

print()
print(f'PASS   : {len(passed)}')
print(f'FAIL   : {len(failed)}')
print(f'SKIP   : {len(skipped)}')
```

```python
# CELL 4: START ALL 17 AGENTS VIA PATHOS
import requests

PATHOS = 'http://localhost:3100'

# Start Pathos if not running
try:
    r = requests.get(PATHOS + '/status', timeout=3)
    data = r.json()
    print(f'PATHOS ONLINE: {data["agents"]} agents')
except:
    print('Starting Pathos...')
    subprocess.Popen(['node', os.path.expanduser('~/pathos/server.js')])
    time.sleep(3)
    try:
        r = requests.get(PATHOS + '/status', timeout=3)
        print('PATHOS STARTED:', r.json())
    except:
        print('Pathos failed to start — check ~/pathos/server.js')

# Dispatch test to each agent
for n, name, role, port in agents:
    try:
        r = requests.post(PATHOS + '/dispatch',
            json={'agent': name, 'prompt': 'status check'},
            timeout=5)
        data = r.json()
        print(f'  ✓ {name:12} — {data.get("response","")[:60]}')
    except Exception as e:
        print(f'  ✗ {name:12} — {str(e)[:40]}')
```
